# Definition of Heat Equation Problem
------------------------------------------------------------------
$
\begin{array}{llll}
    \partial_t u - \langle \nabla, a\nabla u \rangle & = & f & \\
    u(x,0) & = & u_0(x), & \ x \in \Omega \\
        u(x,t) & = & g(x, t), & \ x \in \partial \Omega
\end{array}
$

# Case 1, a) simple 1D, time independent example
------------------------------------------------------------------
Used Parameters:

$ \sigma = 0.03,\quad \mu = 0.5$

Domain 

$
x \in I = [0, 1]
$


$
u(x) = x(1 - x) \quad \Rightarrow \quad u_{x} = 1 - 2x, \quad u_{t} = 0 
$

$
a(x) = 1 + e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$

$
\begin{array}{llll}
    \langle \nabla, a\nabla u \rangle & = & \frac{\partial}{\partial x} \left( (1+e^{-\frac{(x - \mu)^2}{2\sigma^2}})(1 - 2x) \right) \\
    & = & -2 - \frac{1}{\sigma^{2}}  \left(2\sigma^{2} + (1 - 2x)(x - \mu)\right) e^{-\frac{(x-\mu)^{2}}{2\sigma^{2}}}
\end{array}
$

$ 
\begin{array}{llll}
    u_t - \langle \nabla, a\nabla u \rangle & = & 0 - \langle \nabla, a\nabla u \rangle \\
    & = & -\left(-2 - \frac{1}{\sigma^{2}}  \left(2\sigma^{2} + (1 - 2x)(x - \mu)\right) e^{-\frac{(x-\mu)^{2}}{2\sigma^{2}}}\right) \\
    & = & 2 + \frac{1}{\sigma^{2}}  \left(2\sigma^{2} + (1 - 2x)(x - \mu)\right) e^{-\frac{(x-\mu)^{2}}{2\sigma^{2}}} \\
    & = & f(x)
\end{array}
$

IC : none

BC: $u(0) = u(1) = 0$

```python
# Case 1, a) simple 1D, time independent example
# ------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt

sigma = 0.05  
mu = 0.5

x = np.linspace(0, 1, 1000)
u = x * (1 - x) # !!! 40* just for plot!!!
a =  1 + np.exp(-((x - mu)**2 / (2*sigma**2)))

numerator = (2*sigma**2 + (1 - 2*x)*(x - mu))
exponent = np.exp(-((x - mu)**2 / (2*sigma**2)))
factor = 1/(sigma**2)
f =  2 + numerator * exponent * factor

plt.figure(figsize=(8, 4))
plt.plot(x, u, label=r"$u(x)$")
plt.plot(x, a, label=r"$a(x)$")
plt.plot(x, f, label=r"$f(x)$")
plt.title('1D Gaussian Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```

# Case 1, a) simple 1D, time dependent example

```python
# Re-import necessary libraries after reset
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec


sigma = 0.05  
mu = 0.5

# Define the functions
def u_time_depend(x_in):
    return np.exp(-x_in[:, 1:]) * x_in[:, 0:1]*(1 - x_in[:, 0:1])

def f_time_depend(x_in):
    f = np.exp(-x_in[:, 1:]) * (1/sigma**2 * ((1 - 2*x_in[:, 0:1])*(x_in[:, 0:1] - mu) + 2*sigma**2) * np.exp(-((x_in[:, 0:1]-mu)**2)/(2*sigma**2)) - x_in[:, 0:1]*(1 - x_in[:, 0:1]) + 2)
    return f    
    
# Prepare inputs for evaluation
# Define test domain [0,2]x[0,10]
X = np.linspace(0, 1, 1000)
T = np.linspace(0, 6, 100)
X_test, T_test = np.meshgrid(X, T)
     
XT = np.column_stack((X_test.flatten(), T_test.flatten()))

# Evaluate functions
u_vals = u_time_depend(XT).reshape(X_test.shape)
f_vals = f_time_depend(XT).reshape(X_test.shape)

# Evaluate a(x)
a = 1 + np.exp(-0.5 * ((X - mu) / sigma)**2)
A = np.tile(a, (T.shape[0], 1))

# Plotting
fig = plt.figure(figsize=(18, 6))
gs = gridspec.GridSpec(1, 3)

# Plot u(x, t)
ax1 = fig.add_subplot(gs[0, 0], projection='3d')
ax1.plot_surface(X_test, T_test, u_vals, cmap='viridis')
ax1.set_xlabel('x')
ax1.set_ylabel('t')
ax1.set_zlabel('u(x,t)')
ax1.set_title(r"$u(x,t)$")

# Plot f(x, t)
ax2 = fig.add_subplot(gs[0, 1], projection='3d')
ax2.plot_surface(X_test, T_test, f_vals, cmap='plasma')
ax2.set_xlabel('x')
ax2.set_ylabel('t')
ax2.set_zlabel('f(x,t)')
ax2.set_title(r"$f(x,t)$")

# Plot a(x)
ax3 = fig.add_subplot(gs[0, 2], projection='3d')
ax3.plot_surface(X_test, T_test, A, cmap='cividis')
ax3.set_xlabel('x')
ax3.set_ylabel('t')
ax3.set_zlabel('a(x)')
ax3.set_title(r"$a(x)$")

plt.tight_layout()
plt.show()
```

$
\begin{array}{llll}

\begin{array}{llll}
    u_t - \langle \nabla, a\nabla u \rangle & = & - \left( - \frac{\pi}{\sigma^2 } \left((x - {\mu_x}) \cos(\pi x) \sin(\pi y) + (y - {\mu_y})\sin(\pi x)\cos(\pi y) + 2\pi{\sigma}^{2}\sin(\pi x)\sin(\pi y) \right) e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}} - 2\pi^2\sin(\pi x) \sin(\pi y) \right) \\

```python
# Re-import libraries after state reset
import numpy as np
import matplotlib.pyplot as plt

# Parameters
sigma = 0.05
sigma_u = 0.2
mu_x = 0.5
mu_y = 0.5

# Define domain
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)

# Function definitions
u = np.sin(np.pi*X) * np.sin(np.pi * Y)

a = 1 + np.exp(-((X - mu_x)**2 + (Y - mu_y)**2) / (2 * sigma**2))

import numpy as np

def func(x, y, mu_x=0.5, mu_y=0.5, sigma=0.1):
    exp_term = np.exp(-((x - mu_x)**2 + (y - mu_y)**2) / (2 * sigma**2))
    term1 = (x - mu_x) * np.cos(np.pi * x) * np.sin(np.pi * y)
    term2 = (y - mu_y) * np.sin(np.pi * x) * np.cos(np.pi * y)
    term3 = 2 * np.pi * sigma**2 * (np.sin(np.pi * x) * np.sin(np.pi * y))
    first_part = (np.pi / sigma**2) * (term1 + term2 + term3) * exp_term
    second_part = 2 * np.pi**2 * np.sin(np.pi * x) * np.sin(np.pi * y)
    return first_part + second_part


f = func(X, Y)

# Plotting the functions in 3D
fig = plt.figure(figsize=(18, 5))

# Plot u(x,y)
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(X, Y, u, cmap='viridis')
ax1.set_title(r'$u(x,y)$')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('u(x,y)')

# Plot f(x,y)
ax2 = fig.add_subplot(132, projection='3d')
ax2.plot_surface(X, Y, f, cmap='plasma')
ax2.set_title(r'$f(x,y)$')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('f(x,y)')

# Plot a(x,y)
ax3 = fig.add_subplot(133, projection='3d')
ax3.plot_surface(X, Y, a, cmap='cividis')
ax3.set_title(r'$a(x,y)$')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_zlabel('a(x,y)')

plt.tight_layout()
plt.show()
```

    &   & - e^{-t} \left( \frac{\pi}{\sigma^2 } \left((y - {\mu_y}) \sin(\pi x) \cos(\pi y) + \pi{\sigma}^{2} \sin(\pi x) \sin(\pi y) \right) e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}} + \pi^2 \sin(\pi x) \sin(\pi y) \right) \\
    & = & - e^{-t} \left( \frac{\pi}{\sigma^2 } \left((x - {\mu_x}) \cos(\pi x) \sin(\pi y) + (y - {\mu_y})\sin(\pi x)\cos(\pi y) + 2\pi{\sigma}^{2}\sin(\pi x)\sin(\pi y) \right) e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}} + 2\pi^2\sin(\pi x) \sin(\pi y) \right)
\end{array}
$

$ 
\begin{array}{llll}
    u_t - \langle \nabla, a\nabla u \rangle & = & - e^{-t}\sin (\pi x) \sin(\pi y)  + e^{-t} \left( \frac{\pi}{\sigma^2 } \left((x - {\mu_x}) \cos(\pi x) \sin(\pi y) + (y - {\mu_y})\sin(\pi x)\cos(\pi y) + 2\pi{\sigma}^{2}\sin(\pi x)\sin(\pi y) \right) e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}} + 2\pi^2\sin(\pi x) \sin(\pi y) \right) \\ 
    & = & e^{-t} \left( \frac{\pi}{\sigma^2 } \left((x - {\mu_x}) \cos(\pi x) \sin(\pi y) + (y - {\mu_y})\sin(\pi x)\cos(\pi y) + 2\pi{\sigma}^{2}\sin(\pi x)\sin(\pi y) \right) e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}} + (2\pi^2 - 1) \sin(\pi x) \sin(\pi y) \right) \\ 
    & = & f(x,y)
\end{array}
$

IC : $u(x,0) = 0$

BC: $u(0) = u(1) = 0$



```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
sigma = 0.05
mu_x = 0.5
mu_y = 0.5
lmbda = 1

# Define spatial domain
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)

# Gaussian diffusivity
def a(x, y):
    return 1 + np.exp(-((x - mu_x) ** 2 + (y - mu_y) ** 2) / (2 * sigma ** 2))

# Solution u(x,y,t)
def u(x, y, t):
    return np.exp(-t) * np.sin(np.pi * x) * np.sin(np.pi * y)

# Source term f(x,y,t)
def f(x, y, t):
    pi = np.pi
    gaussian_exponent = -((x - mu_x) ** 2 + (y - mu_y) ** 2) / (2 * sigma ** 2)
    gaussian_term = np.exp(gaussian_exponent)
    first_term = (x - mu_x) * np.cos(pi * x) * np.sin(pi * y)
    second_term = (y - mu_y) * np.sin(pi * x) * np.cos(pi * y)
    third_term = 2 * pi * sigma ** 2 * np.sin(pi * x) * np.sin(pi * y)
    combined_term = first_term + second_term + third_term
    exponential_decay = np.exp(-t)
    sinusoidal_term = (2 * pi ** 2 - 1) * np.sin(pi * x) * np.sin(pi * y)
    result = exponential_decay * ((pi / sigma ** 2) * combined_term * gaussian_term + sinusoidal_term)
    return result

# Plotting for selected time instances
times = [0, 2, 4, 6]
fig = plt.figure(figsize=(16, 12))

for i, t in enumerate(times):
    # u(x,y,t)
    ax1 = fig.add_subplot(4, 2, 2*i+1, projection='3d')
    u_values = u(X, Y, t)
    ax1.plot_surface(X, Y, u_values, cmap='viridis')
    ax1.set_title(f'$u(x,y,t)$ at $t={t}$')
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    ax1.set_zlabel('u')

    # f(x,y,t)
    ax2 = fig.add_subplot(4, 2, 2*i+2, projection='3d')
    f_values = f(X, Y, t)
    ax2.plot_surface(X, Y, f_values, cmap='plasma')
    ax2.set_title(f'$f(x,y,t)$ at $t={t}$')
    ax2.set_xlabel('x')
    ax2.set_ylabel('y')
    ax2.set_zlabel('f')

plt.tight_layout()
plt.show()

# Plot a(x,y) separately (time-independent)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
a_values = a(X, Y)
ax.plot_surface(X, Y, a_values, cmap='coolwarm')
ax.set_title('$a(x,y)$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('a')
plt.show()
```
