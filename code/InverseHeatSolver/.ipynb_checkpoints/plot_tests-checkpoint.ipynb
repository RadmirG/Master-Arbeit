# Definition of Heat Equation Problem
------------------------------------------------------------------
$
\begin{array}{llll}
    \partial_t u - \langle \nabla, a\nabla u \rangle & = & f & \\
    u(x,0) & = & u_0(x), & \ x \in \Omega \\
        u(x,t) & = & g(x, t), & \ x \in \partial \Omega
\end{array}
$

# Case 1, a) simple 1D, time independent example
------------------------------------------------------------------
Used Parameters:

$ \sigma = 0.03,\quad \mu = 0.5$

Domain 

$
x \in I = [0, 1]
$


$
u(x) = x(1 - x) \quad \Rightarrow \quad u_{x} = 1 - 2x, \quad u_{t} = 0 
$

$
a(x) = \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$

$
\begin{array}{llll}
    \langle \nabla, a\nabla u \rangle & = & \frac{\partial}{\partial x} \left( \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}\left(1 - 2x\right) \right) \\
    & = & \frac{1}{\sigma^3 \sqrt{2 \pi}}\left(2x^{2} - \left(2{\mu} + 1\right) x - 2{\sigma}^{2} + {\mu}\right) e^{-\frac{\left(x - {\mu}\right)^{2}}{2{\sigma}^{2}}}
\end{array}
$

$ 
\begin{array}{llll}
    u_t - \langle \nabla, a\nabla u \rangle & = & 0 - \langle \nabla, a\nabla u \rangle \\
    & = & \frac{1}{\sigma^3 \sqrt{2 \pi}}\left(2x^{2} - \left(2{\mu} + 1\right) x - 2{\sigma}^{2} + {\mu}\right) e^{-\frac{\left(x - {\mu}\right)^{2}}{2{\sigma}^{2}}} \\
    & = & f(x)
\end{array}
$

IC : none

BC: $u(0) = u(1) = 0$

```python
# Case 1, a) simple 1D, time independent example
# ------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt

sigma = 0.05  
mu = 0.5

x = np.linspace(0, 1, 1000)
u = 40*x * (1 - x) # !!! 40* just for plot!!!
a = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu)**2 / (2*sigma**2)))

numerator = (2 * x**2 - (2 * mu + 1) * x - 2 * sigma**2 + mu)
exponent = np.exp(-((x - mu)**2 / (2*sigma**2)))
factor = 1/(sigma**3 * np.sqrt(2*np.pi))
f =  -numerator * exponent * factor

plt.figure(figsize=(8, 4))
plt.plot(x, u, label=r"$u(x)$")
plt.plot(x, a, label=r"$a(x)$")
plt.plot(x, f, label=r"$f(x)$")
plt.title('1D Gaussian Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```

# Case 1, b) 1D, time independent example

Used Parameters:

$ \sigma = 0.03,\quad \mu = 1 $

Domain 

$
x \in I = [0, 1]

$

$

$

    & = & - \frac{\pi}{\sigma^{3}\sqrt{2 \pi}} \left(\left(x - {\mu} \right) \cos\left(\pi x\right) + \pi{\sigma}^{2} \sin\left(\pi x\right)\right)  e^{-\frac{(x - \mu)^2}{2\sigma^2}}

$

$ 
\begin{array}{llll}
    u_t - \langle \nabla, a\nabla u \rangle & = & 0 - \langle \nabla, a\nabla u \rangle \\

    & = & \frac{\pi}{\sigma^{3}\sqrt{2 \pi}} \left(\left(x - {\mu}\right) \cos\left(\pi x\right) + \pi{\sigma}^{2} \sin\left(\pi x\right)\right)  e^{-\frac{(x - \mu)^2}{2\sigma^2}} \\
    & = & f(x)
\end{array}
$

IC : none

BC: $u(0) = u(2) = 0$

```python
# Case 1, b) little bit complicated 1D, time independent example
# ------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt

sigma = 0.05  
mu = 0.5

x = np.linspace(0, 1, 1000)
u = np.sin(np.pi*x) # just for plot

a = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu)**2 / (2*sigma**2)))

numerator = ((x-mu)*np.cos(np.pi*x) - np.pi * sigma**2 * np.sin(np.pi*x))

exponent = np.exp(-((x - mu)**2 / (2*sigma**2)))

factor = np.pi / (sigma**3 * np.sqrt(2*np.pi))

f =  factor * numerator * exponent

plt.figure(figsize=(8, 4))
plt.plot(x, u, label=r"$u(x)$")
plt.plot(x, a, label=r"$a(x)$")
plt.plot(x, f, label=r"$f(x)$")
plt.title('1D Gaussian Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```

# Case 2) 1D, time dependent example
------------------------------------------------------------------

$ \sigma = 0.05, \quad \mu = 0, \quad x_0 = 1 $


$
x \in I = [0, 1], \quad t \in T = [0,6]
$


$
u(x) = e^{-\lambda t}\sin (\pi x), \quad \lambda = 1 \quad \Rightarrow \quad u_{x} = \pi e^{-t} \cos(\pi x), \quad u_{t} = -e^{-t} \sin(\pi x) 
$

$
a(x) = \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$

$
\begin{array}{llll}
    \langle \nabla, a\nabla u \rangle & = & \frac{\partial}{\partial x} \left( \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}\left(\pi e^{-t} \cos(\pi x)\right) \right) \\
    & = & -\frac{\pi}{{\sigma}^{3}\sqrt{2\pi} } \left(\pi{\sigma}^{2} \sin\left(\pi x\right) + \left(x - {\mu} \right) \cos\left(\pi x\right)\right) e^{-\frac{\left(x - {\mu}\right)^{2}}{2{\sigma}^{2}} - t}
\end{array}
$

$ 
\begin{array}{llll}
    u_t - \langle \nabla, a\nabla u \rangle & = & -e^{-t} \sin(\pi x) - \left( -\frac{\pi}{{\sigma}^{3}\sqrt{2\pi} } \left(\pi{\sigma}^{2} \sin\left(\pi x\right) + \left(x - {\mu} \right) \cos\left(\pi x\right)\right) e^{-\frac{\left(x - {\mu}\right)^{2}}{2{\sigma}^{2}} - t}\right) \\ 
    & = & \left( \frac{\pi}{{\sigma}^{3}\sqrt{2\pi} } \left(\pi{\sigma}^{2} \sin\left(\pi x\right) + \left(x - {\mu}\right) \cos\left(\pi x\right)\right) e^{-\frac{\left(x - {\mu}\right)^{2}}{2{\sigma}^{2}}} - \sin(\pi x) \right) e^{-t}\\ 
    & = & f(x)
\end{array}
$

IC : $u(x,0) = \sin(\pi x)$

BC: $u(0) = u(2) = 0$


```python
# Re-import necessary libraries after reset
import numpy as np
import matplotlib.pyplot as plt


sigma = 0.05  
mu = 0.5

# Define the functions
def u_time_depend(x_in):
    return 1 + np.exp(-x_in[:, 1:]) * np.sin(np.pi * x_in[:, 0:1])

def f_time_depend(x_in):
    factor = - np.pi / (sigma**2 * np.sqrt(2*np.pi))
    e_x_dpnt = np.exp(-((x_in[:, 0:1] - mu)**2 / (2*sigma**2)))
    e_t_dpnt = np.exp(-x_in[:, 1:])
    inner_par = np.pi * sigma**2 * np.sin(np.pi*x_in[:, 0:1]) + (x_in[:, 0:1] - mu)*np.cos(np.pi*x_in[:, 0:1]) 
    
    f = (factor*inner_par*e_x_dpnt - np.sin(np.pi * x_in[:, 0:1]))*e_t_dpnt
    
    return f

# Prepare inputs for evaluation
# Define test domain [0,2]x[0,10]
X = np.linspace(0, 1, 1000)
T = np.linspace(0, 6, 100)
X_test, T_test = np.meshgrid(X, T)
     
XT = np.column_stack((X_test.flatten(), T_test.flatten()))

# Evaluate functions
u_vals = u_time_depend(XT).reshape(X_test.shape)
f_vals = f_time_depend(XT).reshape(X_test.shape)

fig = plt.figure(figsize=(10, 8))

# Plot observed u(x, t)
ax1 = fig.add_subplot(221, projection='3d')
ax1.plot_surface(X_test, T_test, u_vals, cmap='viridis')
ax1.set_title(r"$u(x,t)$")
ax1.set_xlabel('x')
ax1.set_ylabel('t')
ax1.set_zlabel('u(x,t)')

# Plotting f(x,t)
ax2 = fig.add_subplot(222, projection='3d')
ax2.plot_surface(X_test, T_test, f_vals, cmap='plasma')
ax2.set_title(r"$f(x,t)$")
ax2.set_xlabel(r'$x$')
ax2.set_ylabel(r'$t$')
ax2.set_zlabel(r'$f(x,t)$')

# Adjust the spacing between subplots
#plt.subplots_adjust(hspace=0, wspace=0)

plt.tight_layout()
plt.show()

a = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((X - mu) / sigma)**2)

# Additional plot for predicted a(x)
plt.figure(figsize=(8, 6))
plt.plot(X, a, label=r"Learned $a(x)$")  # Time-independent
plt.xlabel(r"$x$")
plt.ylabel(r"$a(x)$")
plt.title(r"$a(x)$")
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Plots with time slider

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Parameters
sigma = 0.05
mu = 0.5

# Define functions
def u_time_depend(x, t):
    return np.exp(-t) * np.sin(np.pi * x)

def f_time_depend(x, t):
    factor = - np.pi / (sigma**2 * np.sqrt(2*np.pi))
    e_x_dpnt = np.exp(-((x - mu)**2 / (2*sigma**2)))
    e_t_dpnt = np.exp(-t)
    inner_par = np.pi * sigma**2 * np.sin(np.pi * x) + (x - mu)*np.cos(np.pi * x)
    return (factor * inner_par * e_x_dpnt - np.sin(np.pi * x)) * e_t_dpnt

# Domain
x = np.linspace(0, 1, 1000)
initial_t = 0

# Initial plots
fig, (ax_u, ax_f) = plt.subplots(2, 1, figsize=(10, 8))
plt.subplots_adjust(bottom=0.25)

line_u, = ax_u.plot(x, u_time_depend(x, initial_t), label=r"$u(x,t)$")
ax_u.set_title(r"$u(x,t)$")
ax_u.set_xlabel("x")
ax_u.set_ylabel("u(x,t)")
ax_u.grid()

line_f, = ax_f.plot(x, f_time_depend(x, initial_t), label=r"$f(x,t)$")
ax_f.set_title("$f(x,t)$")
ax_f.set_xlabel("x")
ax_f.set_ylabel("f(x,t)")
ax_f.grid()

# Slider setup
ax_slider = plt.axes([0.25, 0.03, 0.65, 0.03])
t_slider = Slider(ax_slider, 'Time', 0, 10, valinit=initial_t)

# Update function
def update(val):
    t = t_slider.val
    line_u.set_ydata(u_time_depend(x, t))
    line_f.set_ydata(f_time_depend(x, t))
    fig.canvas.draw_idle()

t_slider.on_changed(update)

#plt.tight_layout()
#fig = plt.figure(figsize=(14, 10), constrained_layout=True)
plt.show()
```

# Case 3) 2D, time independent example
------------------------------------------------------------------

$ \sigma = 0.05, \quad \mu_x = 0.5, \quad \mu_y = 0.5 $


$
(x,y) \in I = [0, 1]\times [0,1]
$


$
u(x,y) = \sin (\pi x) \sin(\pi y) \quad \Rightarrow \quad \nabla u(x,y) = \begin{pmatrix}\pi \cos(\pi x) \\ \pi \cos(\pi y)\end{pmatrix}, \quad \quad u_{t} = 0
$

$
a(x,y) = \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}}
$

$
\begin{array}{llll}
    \langle \nabla, a\nabla u \rangle & = & \nabla \left( \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}} \begin{pmatrix}\pi \cos(\pi x) \\ \pi \cos(\pi y)\end{pmatrix} \right)\\
    & = & \frac{\partial}{\partial x} \left( \frac{\pi}{\sigma \sqrt{2 \pi}} e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}} \cos(\pi x) \right) \\
    & + & \frac{\partial}{\partial y} \left( \frac{\pi}{\sigma \sqrt{2 \pi}} e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}} \cos(\pi y) \right) \\
    & = & - \frac{\pi}{{\sigma}^{3}\sqrt{2\pi} } \left(\pi{\sigma}^{2} \sin(\pi x) + (x - {\mu_x}) \cos(\pi x)\right) e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}} \\
    &   & - \frac{\pi}{{\sigma}^{3}\sqrt{2\pi} } \left(\pi{\sigma}^{2} \sin(\pi y) + (y - {\mu_y}) \cos(\pi y)\right) e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}} \\
    & = & - \frac{\pi}{{\sigma}^{3}\sqrt{2\pi} } \left( \pi{\sigma}^{2}( \sin(\pi x) + \sin(\pi y)) + (x - {\mu_x}) \cos(\pi x) + (y - {\mu_y}) \cos(\pi y) \right) e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}}
\end{array}
$

$ 
\begin{array}{llll}
    u_t - \langle \nabla, a\nabla u \rangle & = & - \left( - \frac{\pi}{{\sigma}^{3}\sqrt{2\pi} } \left( \pi{\sigma}^{2}( \sin(\pi x) + \sin(\pi y)) + (x - {\mu_x}) \cos(\pi x) + (y - {\mu_y}) \cos(\pi y) \right) e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}} \right) \\ 
    & = & \frac{\pi}{{\sigma}^{3}\sqrt{2\pi} } \left( \pi{\sigma}^{2}( \sin(\pi x) + \sin(\pi y)) + (x - {\mu_x}) \cos(\pi x) + (y - {\mu_y}) \cos(\pi y) \right) e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}}\\ 
    & = & f(x,y)
\end{array}
$

IC : $u(x,0) = 0$

BC: $u(0) = u(2) = 0$


```python
# Re-import libraries after state reset
import numpy as np
import matplotlib.pyplot as plt

# Parameters
sigma = 0.05
mu_x = 0.5
mu_y = 0.5

# Define domain
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)

# Function definitions
u = np.sin(np.pi * X) * np.sin(np.pi * Y)

a = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((X - mu_x)**2 + (Y - mu_y)**2) / (2 * sigma**2))

f = (np.pi / (sigma**3 * np.sqrt(2 * np.pi))) * (
    np.pi * sigma**2 * (np.sin(np.pi * X) + np.sin(np.pi * Y)) +
    (X - mu_x) * np.cos(np.pi * X) +
    (Y - mu_y) * np.cos(np.pi * Y)
) * np.exp(-((X - mu_x)**2 + (Y - mu_y)**2) / (2 * sigma**2))

# Plotting the functions in 3D
fig = plt.figure(figsize=(18, 5))

# Plot u(x,y)
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(X, Y, u, cmap='viridis')
ax1.set_title(r'$u(x,y) = \sin(\pi x) \sin(\pi y)$')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('u(x,y)')

# Plot a(x,y)
ax2 = fig.add_subplot(132, projection='3d')
ax2.plot_surface(X, Y, a, cmap='plasma')
ax2.set_title(r'$a(x,y) = \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}}$')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('a(x,y)')

# Plot f(x,y)
ax3 = fig.add_subplot(133, projection='3d')
ax3.plot_surface(X, Y, f, cmap='inferno')
ax3.set_title(r'$f(x,y)$')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_zlabel('f(x,y)')

plt.tight_layout()
plt.show()
```

# Case 4) 2D, time dependent example
------------------------------------------------------------------

$ \sigma = 0.05, \quad \mu_x = 0.5, \mu_y = 0.5 $


$
x \in I = [0, 1]\times[0,1], \quad t \in T = [0,6]
$

$
u(x,y) = e^{-\lambda t}\sin (\pi x) \sin(\pi y), \quad \lambda = 1,  \quad \Rightarrow \quad \nabla u(x,y) = e^{-t} \begin{pmatrix}\pi \cos(\pi x) \\ \pi \cos(\pi y)\end{pmatrix}, \quad \quad u_{t} = - e^{-t}\sin (\pi x) \sin(\pi y)
$

$
a(x,y) = \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}}
$

$
\begin{array}{llll}
    \langle \nabla, a\nabla u \rangle & = & \nabla \left( \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}} e^{-t}  \begin{pmatrix}\pi \cos(\pi x) \\ \pi \cos(\pi y)\end{pmatrix} \right)\\
    & = & \frac{\partial}{\partial x} \left( \frac{\pi}{\sigma \sqrt{2 \pi}} e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}-t} \cos(\pi x) \right) \\
    & + & \frac{\partial}{\partial y} \left( \frac{\pi}{\sigma \sqrt{2 \pi}} e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}-t} \cos(\pi y) \right) \\
    & = & - \frac{\pi}{{\sigma}^{3}\sqrt{2\pi} } \left(\pi{\sigma}^{2} \sin(\pi x) + (x - {\mu_x}) \cos(\pi x)\right) e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}-t} \\
    &   & - \frac{\pi}{{\sigma}^{3}\sqrt{2\pi} } \left(\pi{\sigma}^{2} \sin(\pi y) + (y - {\mu_y}) \cos(\pi y)\right) e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}-t} \\
    & = & - \frac{\pi}{{\sigma}^{3}\sqrt{2\pi} } \left( \pi{\sigma}^{2}( \sin(\pi x) + \sin(\pi y)) + (x - {\mu_x}) \cos(\pi x) + (y - {\mu_y}) \cos(\pi y) \right) e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}-t}
\end{array}
$

$ 
\begin{array}{llll}
    u_t - \langle \nabla, a\nabla u \rangle & = & - e^{-t}\sin (\pi x) \sin(\pi y) - \left( - \frac{\pi}{{\sigma}^{3}\sqrt{2\pi} } \left( \pi{\sigma}^{2}( \sin(\pi x) + \sin(\pi y)) + (x - {\mu_x}) \cos(\pi x) + (y - {\mu_y}) \cos(\pi y) \right) e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}-t} \right) \\ 
    & = & e^{-t} \left( \frac{\pi}{{\sigma}^{3}\sqrt{2\pi} } \left( \pi{\sigma}^{2}( \sin(\pi x) + \sin(\pi y)) + (x - {\mu_x}) \cos(\pi x) + (y - {\mu_y}) \cos(\pi y) \right) e^{-\frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2\sigma^2}} - \sin (\pi x) \sin(\pi y) \right) \\ 
    & = & f(x,y)
\end{array}
$

IC : $u(x,0) = \sin(\pi x)\sin(\pi y)$

BC: $u(0) = u(2) = 0$


```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
sigma = 0.05
mu_x = 0.5
mu_y = 0.5
lmbda = 1

# Define spatial domain
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)

# Gaussian diffusivity
def a(x, y):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu_x)**2 + (y - mu_y)**2) / (2 * sigma**2))

# Solution u(x,y,t)
def u(x, y, t):
    return np.exp(-lmbda * t) * np.sin(np.pi * x) * np.sin(np.pi * y)

# Source term f(x,y,t)
def f(x, y, t):
    gaussian_part = (np.pi / (sigma**3 * np.sqrt(2 * np.pi))) * (
        np.pi * sigma**2 * (np.sin(np.pi * x) + np.sin(np.pi * y)) +
        (x - mu_x) * np.cos(np.pi * x) +
        (y - mu_y) * np.cos(np.pi * y)
    ) * np.exp(-((x - mu_x)**2 + (y - mu_y)**2) / (2 * sigma**2))
    return np.exp(-t) * (gaussian_part - np.sin(np.pi * x) * np.sin(np.pi * y))

# Plotting for selected time instances
times = [0, 2, 4, 6]
fig = plt.figure(figsize=(16, 12))

for i, t in enumerate(times):
    # u(x,y,t)
    ax1 = fig.add_subplot(4, 2, 2*i+1, projection='3d')
    u_values = u(X, Y, t)
    ax1.plot_surface(X, Y, u_values, cmap='viridis')
    ax1.set_title(f'$u(x,y,t)$ at $t={t}$')
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    ax1.set_zlabel('u')

    # f(x,y,t)
    ax2 = fig.add_subplot(4, 2, 2*i+2, projection='3d')
    f_values = f(X, Y, t)
    ax2.plot_surface(X, Y, f_values, cmap='plasma')
    ax2.set_title(f'$f(x,y,t)$ at $t={t}$')
    ax2.set_xlabel('x')
    ax2.set_ylabel('y')
    ax2.set_zlabel('f')

plt.tight_layout()
plt.show()

# Plot a(x,y) separately (time-independent)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
a_values = a(X, Y)
ax.plot_surface(X, Y, a_values, cmap='coolwarm')
ax.set_title('$a(x,y)$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('a')
plt.show()
```
