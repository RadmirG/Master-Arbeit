<a id="s0"></a><h2 style="color:rgb(127,203,223)">4. Unüberwachtes Lernen</h2>

In diesem Kapitel wollen wir uns mit ausgewählten Methoden des unüberwachten Lernens beschäftigen. Neben der Dimensionsreduktion von Merkmalsvektoren am Beispiel der Hauptkomponentenanalyse (PCA) werden wir uns mit zwei ausgewählten Verfahren aus dem Bereich der Clusteranalyse beschäftigen. Bei letzterer geht es darum, gegebene Merkmale automatisiert so zu gruppieren, dass Objekte einer Gruppe untereinander hinsichtlich spezifischer Eigenschaften ähnlicher sind als Merkmale unterschiedlicher Gruppen. Da hier keine Form von Label (Ground Truth) vorausgesetzt bzw. benötigt wird, sprechen wir hier von unüberwachtem Lernen (unsupervised learning).

<a id="s1"></a><h3 style="color:rgb(127,203,223)">4.1 Principal Component Analysis</h3>

Die Hauptkomponentenanalyse (<b>P</b>rincipal <b>C</b>omponent <b>A</b>nalysis) ist ein lineares Verfahren zur Dimensionsreduktion. Dimensionsreduktion ist hierbei wie folgt zu verstehen. In der Bildverarbeitung haben wir es häufig mit hochdimensionalen Merkmalsvektoren zu tun. Gegebenenfalls stellt das Bild selbst den Merkmalsvektor dar (siehe auch Beispiel weiter unten). In diesen Fällen ist es sinnvoll, den Merkmalsvektor zunächst so zu transformieren, dass seine Dimension deutlich reduziert wird bei gleichzeitiger Erhaltung seiner Information. Ein lineares Standardverfahren hierzu stellt die so genannte PCA dar. Wir wollen uns zunächst die Grundidee mit Hilfe eines GeoGebra-Applets veranschaulichen, bevor wir in die mathematischen Grundlagen einsteigen:

```python
from IPython.display import IFrame
IFrame(src='https://www.geogebra.org/m/ehwts3r4', width=800, height=600)
```

Nachdem wir, so nötig, die Daten mittelwertfrei gemacht haben, der Ursprung des Koordinatensystems stellt dann automatisch auch den Mittelwert der Punktwolke dar, suchen wir die Ursprungsgerade, bzgl. der die Varianz der auf die Gerade projizierten Punkte maximal wird. Der Vektor, der vom Ursprung aus abgetragen auf einen Punkt zeigt, lässt sich also in einen Anteil parallel und einen Anteil orthogonal zur Geraden zerlegen. Wir merken uns den parallelen Anteil sowie die Gerade und verschieben den orthogonalen Anteil so, dass der Vektor im Ursprung beginnt. Nachdem wir dies für alle Punkte durchgeführt haben, suchen wir erneut die Gerade, bzgl. der die Varianz maximiert wird. Auf diesem Wege erhalten wir eine Reihe von Achsen, die jeweils durch einen Einheitsvektor beschrieben werden können, den wir im Ursprung abtragen. Alle diese Vektoren bilden final eine neue Basis des Vektorraumes. Der erste Basisvektor "berücksichtigt" nun die maximale Varianz in den Daten. Der zweite Basisvektor entsprechend die maximale Varianz bzgl. aller Richtungen orthogonal zur ersten Richtung etc. Unter Verwendung aller Basisvektoren erhalten wir einfach eine neue aber gleichwertige Koordinatendarstellung aller Punkte. Die ursprünglichen Koordinaten können als Koordinaten bzgl. der kanonischen Basis aufgefasst werden.<br>
Die Dimensionsreduktion erfolgt nun einfach dadurch, dass wir nicht alle Basisvektoren für die Rekonstruktion der ursprünglichen Punkte (Ortsvektoren) verwenden sondern stattdessen lediglich die, die zur größten Varianz gehören. Die PCA erzeugt also eine Dimensionsreduktion unter Beibehaltung einer möglichst maximalen Varianz. Die implizite Annahme besteht also darin, dass die Varianz ein Maß für den Informationsgehalt darstellt. Dies kann, muss aber nicht immer gelten. Schauen wir uns nun an, wie wir die entsprechenden Achsen mathematisch bestimmen können.

Wir gehen zunächst davon aus, dass wir die mittelwertbefreiten Merkmalsvektoren $\underline{x}_1,\dots,\underline{x}_N$ (es gilt also $\sum\limits_{i=1}^N \underline{x}_i = \underline{0}$) zu einer Matrix <br><br>
$$\underline{\underline{X}}=\left(\begin{array}{c}  \underline{x}_1^T \\ \vdots \\ \underline{x}_N^T\end{array}\right)$$<br><br>
zusammenfassen. Besteht jeder Merkmalsvektor aus $M$ einzelnen Merkmalen, so besitzt die Matrix $\underline{\underline{X}}$ folglich die Dimension $NxM$. <br>
<p style="text-align: center"><img src="pics/Gluehbirne.jpg" alt="Bild" style="height: 220px; width: 300px;"><br></p><br>
    <p style="text-align: center"><b>Welchen Rang hat diese Matrix?</b></p><br>
Beschreiben wir die Richtung einer ausgewählten Ursprungsgeraden nun durch den Einheitsvektor $\underline{w}$, so erhalten wir für die Varianz der auf die Achse projizierten Punkte den folgenden Ausdruck:<br><br>
$$Var(\underline{w})=\sum\limits_{i=1}^N \left(<\underline{x}_i | \underline{w}>\right)^2 \hspace{2cm}(*)$$<br><br>
Bedenken Sie, dass sich die Varianz wie folgt berechnen lässt: $Var=\frac{1}{N}\sum\limits_{i=1}^N (x_i-\mu)^2$. In diesem Fall ist jedoch der Mittelwert $\mu$ gleich Null. Den Faktor $N$ können wir, da er lediglich eine Konstante darstellt bei der weiteren Rechnung vernachlässigen.<br>
Wir formen nun den Ausdruck (*) wie folgt um:<br><br>
$$Var(\underline{w})=||\underline{\underline{X}}\,\underline{w}||_2^2 = <\underline{\underline{X}}\,\underline{w}|\underline{\underline{X}}\,\underline{w}>=<\underline{w}|\underline{\underline{X}}^T\,\underline{\underline{X}}\,\underline{w}> \hspace{2cm}(**)$$<br><br>
Die Matrix $\underline{\underline{X}}^T\,\underline{\underline{X}}$ entspricht - wieder bis auf den Skalierungsfaktor - der so genannten Kovarianzmatrix. Überzeugen Sie sich davon, dass es sich hierbei um eine symmetrische Matrix handelt. In der linearen Algebra haben Sie gelernt (Spektralsatz für selbstadjungierte Endomorphismen endlichdimensionaler Vektorräume), dass Sie eine symmetrische Matrix stets diagonalisieren können. Anders ausgedrückt, existiert eine Orthonormalbasis aus Eigenvektoren der Matrix $\underline{\underline{X}}^T\,\underline{\underline{X}}$:<br><br>
$$\left(\underline{w}_1,\dots,\underline{w}_M\right) \mbox{ mit } \underline{\underline{X}}^T\,\underline{\underline{X}}\,\underline{w}_i = \sigma_i \, \underline{w}_i \mbox{  mit }\sigma_1 \geq \sigma_2 \geq \cdots \geq \sigma_M$$
Die $\sigma_i$ entsprechen hierbei den Eigenwerten. Das bedeutet, dass wir unseren gesuchten Einheitsvektor $\underline{w}$ aus Gleichung (*) in dieser Basis entwicklen können:<br><br>
$$\underline{w}=\alpha_1\,\underline{w}_1+\cdots+\alpha_M\,\underline{w}_M$$<br><br>
Setzen wir dies nun in die Gleichung (**) ein und berücksichtigen wir, dass die Basisvektoren orthonormiert sind, so erhalten wir:<br><br>
$$\begin{eqnarray}
<\underline{w}|\underline{\underline{X}}^T\,\underline{\underline{X}}\,\underline{w}>&=&<\alpha_1\,\underline{w}_1+\cdots+\alpha_M\,\underline{w}_M|\underline{\underline{X}}^T\,\underline{\underline{X}}(\alpha_1\,\underline{w}_1+\cdots+\alpha_M\,\underline{w}_M)>\\
&=&<\alpha_1\,\underline{w}_1+\cdots+\alpha_M\,\underline{w}_M|\alpha_1\,\sigma_1\,\underline{w}_1+\cdots+\alpha_M\,\sigma_M\,\underline{w}_M>\\
&=&\alpha_1^2\,\sigma_1+\cdots+\alpha_M^2\,\sigma_M\\
&\leq& \sigma_1\,(\alpha_1^2+\cdots+\alpha_M^2)=\sigma_1
\end{eqnarray}$$<br><br>
Wir können unmittelbar erkennen, dass der Eigenvektor $\underline{w}_1$, der zum größten Eigenwert der Kovarianzmatrix gehört, die gesuchte erste Ursprungsgerade beschreibt. Die Varianz selbst bzgl. dieser Geraden ist dann - abgesehen vom Skalierungsfaktor - durch eben diesen Eigenwert $\sigma_1$ gegeben.<br>
Völlig analog verfahren wir bei der Berechnung der noch folgenden Ursprungsgeraden bzw. Achsen. Diese werden dann ganz offensichtlich gerade durch die weiteren Eigenvektoren (der Größe der zugehörigen Eigenwerte nach sortiert) beschrieben. Auf diese Weise erhalten wir, wie bereits oben erwähnt, ein neues Koordinatensystem, welches von den Eigenvektoren der Kovarianzmatrix aufgespannt wird. wir können damit jeden Datenvektor $\underline{x}_i$ nun wie folgt darstellen:<br><br>
$$\underline{x_i} = x_1\,\underline{e}_1+\cdots+x_M\,\underline{e}_M=<\underline{x}_i|\underline{w}_1>\,\underline{w}_1+\cdots+<\underline{x}_i|\underline{w}_M>\,\underline{w}_M$$<br><br>
Hierbei beschreiben die $\underline{e}_1,\cdots,\underline{e}_M$ die kanonischen Einheitsvektoren und die $<\underline{x}_i|\underline{w}_1>,\cdots,<\underline{x}_i|\underline{w}_M>$ stellen die so genannten <b>Principal Components</b> dar. Ob wir nun mit den ursprünglichen Koordinaten $(x_1,\dots,x_M)$ oder den Principal Components rechnen, ist egal, da beide dieselbe Information kodieren. Die gewünschte Dimensionsreduktion erreichen wir nun damit, dass wir statt allen Principal Components nur die verwenden, die zu den größten Varianzen gehören.<br><br>
Schauen wir uns nun ein Beispiel in sklearn an. Hierzu verwenden wir den Iris-Datensatz, da hier 4 verschiedene Merkmale verwendet werden. Durch die PCA können wir uns die Daten projiziert in den dreidimensionalen Raum anschauen:

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from ipywidgets import interactive, fixed
import ipywidgets as widgets

def performPCA(elev, azim):
    
    # load dataset
    data= load_iris()

    # separate features from class labels
    X = data.data
    y = data.target

    # let's perform a PCA and only use the first three principal components
    pca = PCA(n_components=3)
    pca.fit(X)

    X_reduced = pca.transform(X)

    # plot principal components
    fig = plt.figure(1, figsize=(8,6))
    ax = fig.add_subplot(111, projection="3d", elev=elev, azim=azim)
    ax.scatter(X_reduced[:,0], X_reduced[:,1], X_reduced[:,2],c=y, cmap=plt.cm.Set1,edgecolor="k",s=40)
    ax.set_title("First three principal components of Iris-datatset")
    ax.set_xlabel("1st PC")
    ax.set_ylabel("2nd PC")
    ax.set_zlabel("3rd PC")
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])

    plt.show()
    
v=interactive(performPCA,elev=(-90,90),azim=(0,360))
v    
```

<a id="s2"></a><h3 style="color:rgb(127,203,223)">4.2 Eigenfaces</h3>

Ein klassische Anwendung der PCA besteht in der Erkennung von Gesichtern. Hierbei wird die PCA dazu verwendet, die Dimension von Gesichtsbildern, die als Merkmalsvektoren aufgefasst werden, zu reduzieren. Die bei diesem Verfahren basierend auf einer hinreichend großen Datenmenge berechneten Hauptachsen (Principal Axes, Eigenvektoren der Kovarianzmatrix) werden dann als Basis verwendet, in der die Gesichtsbilder entwickelt werden. Jedes einzelne von Ihnen kann dann in den neuen Koordinaten, die nichts anderes als die Principal Components darstellen, kodiert werden. Basierend auf dieser effizienteren Kodierung kann dann eine Erkennung unter Verwendung eines entsprechenden Klassifikationsverfahrens erfolgen. Schauen wir uns das Ganze mal unter Verwendung von sklearn an. Hierzu wird dort eine entsprechende Datenbank zur Verfügung gestellt: 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces
from sklearn.decomposition import PCA
from ipywidgets import interactive, fixed
import ipywidgets as widgets

def eigenfaces(numpcs=30):
    # load olivetti dataset
    X,y = fetch_olivetti_faces(return_X_y=True)
    pca = PCA(n_components=numpcs)
    pca.fit(X)

    # determine principal axes and reshape them to fit the size of images
    eigenfaces = pca.components_.reshape((numpcs,64,64))

    # use colormap gray for plotting images
    plt.set_cmap('gray')

    # print first 4 eigenfaces
    fig, (ax1,ax2,ax3,ax4) = plt.subplots(1,4)
    fig.suptitle("The first four eigenfaces")
    fig.set_size_inches(12,3)

    ax1.set_xticks([])
    ax1.set_yticks([])
    ax2.set_xticks([])
    ax2.set_yticks([])
    ax3.set_xticks([])
    ax3.set_yticks([])
    ax4.set_xticks([])
    ax4.set_yticks([])

    ax1.imshow(eigenfaces[0,:,:])
    ax2.imshow(eigenfaces[1,:,:])
    ax3.imshow(eigenfaces[2,:,:])
    ax4.imshow(eigenfaces[3,:,:])

    plt.show()

    # reconstruct the first image by using only the first numpcs principal axes
    pcs = pca.transform(X[0:1,:])
    img_reduced  = (pca.inverse_transform(pcs)).reshape((64,64))
    img_original = X[0,:].reshape((64,64))
    img_diff = np.abs(img_reduced-img_original)

    # print both, original image and image reconstructed using the first numpcs principal components
    fig, (ax1,ax2,ax3) = plt.subplots(1,3)

    fig.suptitle("Orginal versus reconstructed face image plus difference image")
    fig.set_size_inches(8,3)

    ax1.set_xticks([])
    ax1.set_yticks([])
    ax2.set_xticks([])
    ax2.set_yticks([])
    ax3.set_xticks([])
    ax3.set_yticks([])


    ax1.imshow(img_original)
    ax2.imshow(img_reduced)
    ax3.imshow(img_diff)


v=interactive(eigenfaces,numpcs=(20,400))
v  
```

Die im vorhergehenden Beispiel verwendete Datenbank besteht aus 400 Gesichtsbildern von 40 verschiedenen Personen (also 10 Bilder pro Person mit einer Auflösung von 64x64 Pixeln). Die Matrix $\underline{\underline{X}}$ hat somit die Dimension $400x4096$. Die zugehörige Kovarianzmatrix besitzt folglich die Dimension $4096x4096$. Da wir jedoch nur $400$ verschiedene Bilder verwenden, berechnet sich der Rang der Kovarianzmatrix maximal zu $399$. Bezüglich vieler Hauptachsen ist die Varianz folglich Null. Um den Rechenaufwand - es geht immerhin darum, die Eigenwertzerlegung einer 4096x4096-Matrix zu bestimmen - zu reduzieren, kann man sich eines Tricks bedienen. Dieser soll im Folgenden mathematisch gezeigt werden:

Statt die Eigenwerte und Eigenvektoren der Matrix $\underline{\underline{X}}^T\,\underline{\underline{X}}$ zu berechnen, berechnen wir für den Fall, dass die Anzahl der Spalten der Matrix $\underline{\underline{X}}$ die der Zeilen deutlich übersteigt stattdessen die Eigenwerte und Eigenvektoren der Matrix $\underline{\underline{X}}\,\underline{\underline{X}}^T$. Für das obige Beispiel bedeutet dies, dass wir die Eigenwertzerlegung "nur" für eine Matrix der Größe $400x400$ statt $4096x4096$ durchführen müssen. Es soll gezeigt werden, dass die Eigenwerte der Matrizen $\underline{\underline{X}}\,\underline{\underline{X}}^T$ und $\underline{\underline{X}}^T\,\underline{\underline{X}}$ gleich sind und die Eigenvektoren (zu Eigenwerten ungleich Null) umgerechnet werden können. Es sei<br><br>
$$\underline{\underline{X}}\,\underline{\underline{X}}^T\,\underline{w} = \sigma \, \underline{w} \hspace{2cm}\mbox{(*)}$$<br><br> 
dann gilt:<br><br>
\begin{eqnarray}
\underline{\underline{X}}^T\left(\underline{\underline{X}}\,\underline{\underline{X}}^T\,\underline{w}\right) &=& \sigma \, \underline{\underline{X}}^T\, \underline{w}\\
\left(\underline{\underline{X}}^T\,\underline{\underline{X}}\right)\,\underline{\underline{X}}^T\,\underline{w}&=&\sigma \, \underline{\underline{X}}^T\, \underline{w}\\
\end{eqnarray}<br><br>
Wir sehen, dass wir einen Eigenvektor der Matrix $\underline{\underline{X}}\,\underline{\underline{X}}^T$ einfach durch Multiplikation mit der Matrix $\underline{\underline{X}}^T$ in einen Eigenvektor der Matrix $\underline{\underline{X}}^T\,\underline{\underline{X}}$ zum gleichen Eigenwert (ungleich Null) überführen können!!!

<a id="s6"></a><h4 style="color:rgb(127,203,223)">4.2.1 Nichtlineare Dimensionsreduktion - Ein Beispiel</h4>
Hier soll noch ein ganz kurzer Ausblick auf das generelle Thema der Dimensionsreduktion gegeben werden. Die PCA, welche zu den linearen Verfahren zählt, wählt den $m$-dimensionalen Unterraum aus, der die verbliebene Varianz maximiert. Dies trägt jedoch insbesondere nichtlinearen Strukturen innerhalb der Daten keine Rechnung. Andere, nichtlineare Verfahren finden Sie bspw. in der Bibliothek <i>sklearn.manifold</i>. Hier soll als ein Beispiel lediglich das t-Distributed Stochastic Neighbour Embedding (<A href="www.wikipedia.org">t-SNE</A>) gezeigt werden. Hier ist das Ziel, die i.A. hochdimensionalen Daten in einen niedrigdimensionalen Raum (in der Regel 2 oder 3D) so abzubilden, dass Punkte (Samples), die im hochdimensionalen Raum benachbart sind, dies auch noch im niedrigdimensionalen Raum sind. Ebenso sollten Punkte, die im hochdimensionalen Raum weit voneinder entfernt sind, diese Eigenschaft auch im niedrigdimensionalen Raum behalten. Wir schauen uns hierzu ein Beispiel an, ohne hier näher auf die dahinter liegende Mathematik einzugehen. Wir verwenden hierzu exemplarisch Bilder, die jeweils eine der Ziffern 0 bis 5 zeigen. Wir stellen hierbei das Verfahren der PCA (Projektion auf 2 Dimensionen) dem des t-SNE gegenüber. Der Code wurde übernommen und angepasst von <a href="https://scikit-learn.org/stable/auto_examples/manifold/plot_lle_digits.html#sphx-glr-auto-examples-manifold-plot-lle-digits-py">Manifold learning on handwritten digits</a>

```python
from sklearn.datasets import load_digits
import numpy as np
from matplotlib import offsetbox
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA

digits = load_digits(n_class=6)
X, y = digits.data, digits.target
n_samples, n_features = X.shape
n_neighbors = 30


fig, axs = plt.subplots(nrows=10, ncols=10, figsize=(6, 6))
for idx, ax in enumerate(axs.ravel()):
    ax.imshow(X[idx].reshape((8, 8)), cmap=plt.cm.binary)
    ax.axis("off")
_ = fig.suptitle("A selection from the 64-dimensional digits dataset", fontsize=16)

def plot_embedding(X, title):
    fig, ax = plt.subplots()
    fig.set_size_inches(20,20)
    X = MinMaxScaler().fit_transform(X)

    for digit in digits.target_names:
        ax.scatter(
            *X[y == digit].T,
            marker=f"${digit}$",
            s=60,
            color=plt.cm.Dark2(digit),
            alpha=0.425,
            zorder=2,
        )
    shown_images = np.array([[1.0, 1.0]])  # just something big
    for i in range(X.shape[0]):
        # plot every digit on the embedding
        # show an annotation box for a group of digits
        dist = np.sum((X[i] - shown_images) ** 2, 1)
        if np.min(dist) < 4e-3:
            # don't show points that are too close
            continue
        shown_images = np.concatenate([shown_images, [X[i]]], axis=0)
        imagebox = offsetbox.AnnotationBbox(
            offsetbox.OffsetImage(digits.images[i], cmap=plt.cm.gray_r), X[i]
        )
        imagebox.set(zorder=1)
        ax.add_artist(imagebox)

    ax.set_title(title)
    ax.axis("off")

data = X
tsne = TSNE(n_components=2,init="pca",learning_rate="auto",n_iter=500,n_iter_without_progress=150,n_jobs=2,random_state=0)
res = tsne.fit_transform(data, y)
pca = PCA(n_components=2)
res2 = pca.fit_transform(X)
plot_embedding(res, 't-SNE')
plot_embedding(res2, 'PCA')
```

<a id="s3"></a><h3 style="color:rgb(127,203,223)">4.3 Clustering</h3>

Unter Clustering oder auch Cluster-Analyse verstehen wir die Aufgabe, gegebene Objekte (in aller Regel handelt es sich um Merkmalsvektoren) so zu gruppieren, dass einander ähnliche Objekte in derselben Gruppe, einander unähnliche Objekte jedoch in unterschiedlichen Gruppen liegen. Hierbei werden lediglich die Eigenschaften der Objekte (Merkmalsvektoren) berücksichtigt. Die einzelnen Gruppen sind dann die so genannten Cluster. Die Zuordnung eines Objekts zu einer Gruppe nennen wir dann Clustering. Clustering wir häufig auch im Bereich des Data-Minings eingesetzt. Beim Data-Mining geht es darum, Strukturen und Muster in einer großen Menge von Daten zu identifizieren (Knowledge Discovery in Databases). Anders als beim überwachten Lernen, wo für die Erstellung eines Klassifikators bereits vorhandene Klassen-Label vorausgesetzt werden, handelt es sich beim Clustering um ein Verfahren, Daten (Merkmalsvektoren) aufgrund ihrer Eigenschaften in Gruppen (Klassen) automatisiert zusammenfassen zu lassen. Wie Sie es bereits aus dem Bereich des überwachten Lernens kennen, gibt es auch im Bereich der Cluster-Analyse eine Vielzahl von unterschiedlichen Algorithmen. Im Folgenden sollen zwei Repräsentanten näher betrachtet werden. Die Cluster-Analyse kann des Weiteren grob in zwei Bereiche unterteilt werden, dem so genannten <i>Hard Clustering</i> und <i>Soft Clustering</i>. Während bei ersterem jedes Objekt genau einem Cluster zugeordnet wird, wird im zweiten Fall für jedes Cluster ein Grad der Zugehörigkeit berechnet. Hierbei handelt es sich häufig um eine Wahrscheinlichkeit dafür, dass ein gegebenes Objekt einem gegebenen Cluster angehört oder nicht angehört. Wir werden für jede der beiden Bereiche ein Beispielverfahren kennenlernen. Zunächst jedoch ein anschauliches Einführungsbeispiel: <br><br>

Hier sind Punkte in der Ebene abgebildet, die wir uns als die Komponenten eines zweidimensionalen Merkmalvektors vorstellen können. Betrachten wir das linke Bild, so können wir grob drei Cluster identifizieren, die von den dunkelgrünen, dunkelblauen und roten Punkten gebildet werden. Auf dem rechten Bild erkennen wir demgegenüber zwei Cluster bestehend aus den roten und dunkelblauen Punkten. Im Allgemeinen sind die Merkmalsvektoren, wie Sie bereits wissen, hochdimensional, so dass wir Sie nicht einfach visualisieren können. Die Verfahren, die wir uns im Folgenden näher anschauen werden, sind in ihrer Anwendung grundsätzlich auch für hochdimensionale Daten geeignet. Des Weiteren besteht natürlich die Möglichkeit, die Daten zunächst unter Verwendung eines Verfahrens der Dimensionsreduktion (siehe bspw. PCA) zu "verkleinern" bevor dann das eigentlich Clustering-Verfahren angewendet wird. <br>
Schauen wir uns zunächst ein Verfahren aus dem Bereich des <i>Hard Clustering</i> an.

<a id="s4"></a><h4 style="color:rgb(127,203,223)">4.3.1 k-Means-Clustering</h4>

Beim k-Means-Clustering, das man auch als Verfahren der so genannten Vektor-Quantisierung bezeichnet, handelt es sich um einen Repräsentanten des Hard Clustering. Die Vektorquantisierung wird u.a. zur Datenkompression verwendet. Hierbei speichert man für ein gegebenes Objekt (Merkmalsvektor) lediglich das Cluster, dem dieses zugehörig ist. Auf diese Weise kann der Speicherbedarf signifikant reduziert werden. Auch für die Klassifizierung kann das Cluster-Verfahren verwendet werden, indem jedem neuen Objekt (Merkmalsvektor) einfach der Cluster (Klasse) zugewiesen wird, der diesem am nächsten liegt. Wir interessieren uns im Folgenden jedoch zunächst einmal für die Cluster-Analyse selbst.<br>
Bei diesem, wie auch dem nächsten Algorithmus zur Cluster-Analyse, wird die Anzahl der zu findenden Cluster vorgegeben. Die Bestimmung der "optimalen" Anzahl an Clustern (Klassen) kann i.A. mittels trial&error oder durch Verwendung verschiedener Heuristiken erfolgen.<br>
Das <b>k-Means-Clustering</b> (konkret die Variante nach Lloyd) erfolgt folgendermassen:<br>
<ol>
    <li>Zufällige Auswahl von $k$ Cluster-Zentren (so genannte <b>means</b>) aus den vorhandenen Trainingsdaten (es gibt auch andere Varianten)</li>
    <li>Weise jeden Merkmalsvektor genau dem Cluster-Zentrum zu, dem er am nächsten liegt (euklidische Distanz). Sollten mehrere gleich weit entfernt liegen, wird ein Cluster-Zentrum ausgewählt.</li>
    <li>Berechne für jedes Cluster einen neuen Mittelwert, indem über die Merkmalsvektoren gemittelt wird, die dem jeweiligen Cluster angehören (arithmetischer Mittelwert)</li>
    <li>Wiederhole die Schritte 2 & 3 so lange, bis ein Abbruchkriterium erreicht worden ist (Anzahl maximaler Iterationen, Veränderung der means etc.).</li>
</ol>
Die so berechneten Cluster-Zentren (means) repräsentieren nun die eigentlichen Cluster. Jedes Merkmal wird nun genau einem Cluster zugeordnet (Kriterium ist der minimale euklidische Abstand zum mean). Neben der euklidischen Distanz sind auch andere Metriken möglich. Hier muss jedoch sichergestellt sein, dass der Algorithmus auch konvergiert. Das Vorgehen des Algorithmus ist in der unten dargestellten Abbildung noch einmal schematisch zusammengefasst:<br><br>
<p style="text-align: center"><img src="pics/kmeans.png" alt="Bild" style="height: 400px; width: 800px;"><br>Quelle: Wikipedia</p><br><br>
Ganz rechts in oben gezeigter Abbildung ist farblich ein so genanntes <b>Voronoi-Diagramm</b> dargestellt. Hierbei ist jeder Punkt der Ebene in der Farbe eingefärbt worden, die der des nächstgelegenen Cluster-Zentrums (means) entspricht. In der folgenden Abbildung ist der Algorithmus noch einmal dynamisch dargestellt:<br><br><br>
<p style="text-align: center"><img src="pics/kmeans_convergence.gif" alt="Bild" style="height: 300px; width: 400px;"><br>Quelle: Wikipedia</p><br><br>
Es liegt nun auf der Hand, dass die Cluster-Zentren direkt zur Klassifikation verwendet werden können. Jedem neuen Merkmal wird hierzu einfach das "Label" des am nächsten gelegenen Cluster-Zentrums zugeordnet.

Für die Berechnung des k-Means-Clustering verwenden Sie bitte die Klasse <i><A href="https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html#sklearn.cluster.KMeans">sklearn.cluster.KMeans</A></i> aus dem Framework <b>sklearn</b>.  

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix 


    
# load dataset
X,y = load_iris(return_X_y=True)
# data splitting (training and test)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.50, random_state=42)


# separate features from class labels
mu_0 = np.mean(X[np.argwhere(y_train==0)],axis=0)
mu_1 = np.mean(X[np.argwhere(y_train==1)],axis=0)
mu_2 = np.mean(X[np.argwhere(y_train==2)],axis=0)
init_val = np.vstack((mu_0,mu_1,mu_2))

kmeans = KMeans(n_clusters=3,init=init_val, n_init=1).fit(X_train)
y_pred = kmeans.predict(X_test)
print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred)) 
```

<a id="s5"></a><h4 style="color:rgb(127,203,223)">4.3.2 Gaussian Mixture Model</h4>

Während das k-Means-Clustering aus dem letzten Unter-Kapitel zum so genannten Hard-Clustering gehört, lernen Sie nun ein Verfahren kennen, welches der Klasse der Soft-Clustering-Verfahren zugerechnet wird. Hierbei wird ein Merkmalsvektor nicht automatisch nur einem Cluster-Zentrum zugewiesen sondern Sie bestimmen vielmehr jeweils eine Wahrscheinlichkeit dafür, dass der gegebene Merkmalsvektor einem der zur Verfügung stehenden Cluster-Zentren zugehörig ist. Bevor wir jedoch konkret auf das <b>Gaussian Mixture Model</b> eingehen, soll zunächst der Begriff des <b>Mixture Models</b> im Allgemeinen erläutert werden.<br>
Unter einem Mixture Model, im Deutschen würden wir von einem gemischten Modell sprechen, verstehen wir ein probabilistisches Modell welches in der Lage ist, unterschiedliche Sub-Populationen in einer Gesamt-Population zu beschreiben. Formal können wir ein solches Modell durch eine mixture distribution (Mischverteilung) beschreiben. Diese Verteilung beschreibt eine Zufallsvariable, die aus mehreren anderen Zufallsvariablen abgeleitet wird. Dies geschieht dadurch, dass zunächst eine der anderen Zufallsvariablen <i>zufällig</i> ausgewählt und deren Realisierung dann dem Wert der ursprünglichen Zufallsvariablen zugewiesen wird. Wir haben es also mit einer Art doppelten Zufallsverteilung zu tun. Zum einen wählen wir zufällig aus einer Menge von Zufallsvariablen eine Zufallsvariable aus und nehmen deren Wert - der ja ebenfalls zufällig ist - dann als Ergebnis. Zur besseren Veranschaulichung betrachten wir ein einfaches Beispiel:<br>
Nehmen wir an, wir hätten zwei Münzen, eine faire (gleiche Wahrscheinlichkeit für Kopf und Zahl) und eine unfaire (hier soll die Wahrscheinlichkeit für Kopf bei $\frac{2}{3}$ und die für Zahl entsprechend bei $\frac{1}{3}$ liegen) Münze. Wir wählen nun in jedem Durchgang per Zufall jeweils mit der Wahrscheinlichkeit $\frac{1}{2}$ die faire oder die unfaire Münze aus. Sie bekommen jedoch nur das Endergebnis mitgeteilt, wissen also nicht, welcher der beiden Münzen geworfen worden ist. Wir erhalten somit eine Gesamtverteilung von Ergebnissen, die aus zwei Sub-Populationen entsprungen sind (faire versus unfaire Münze). Die Verteilung der Ergebnisse entspricht somit einer gemischten Verteilung und die Ergebnisse werden daher von einem gemischten Modell erzeugt. Für die Verteilung der einzelnen zur Verfügung stehenden Zufallsvariablen (im Beispiel sind es die beiden Münzen) als auch für deren Dimension sind generell diverse Variationen möglich (Binomialverteilung, Gauß-Verteilung, univariat, multivariat,...).

<p style="text-align: center"><img src="pics/Exercise.png" alt="Bild" style="height: 220px; width: 300px;"><br></p><br>
<b><u>Eine kleine Übung zwischendurch...</b></u><P>
    Mit welcher Wahrscheinlichkeit fällt Kopf im obigen Beispiel?

Wir wollen uns hier ausschließlich mit dem <b>G</b>aussian <b>M</b>ixture <b>M</b>odel beschäftigen, gehen also von i.A. mehrdimensionalen kontinuierlichen Merkmalsvektoren aus. In diesem Fall sieht das GMM wie folgt aus:<br>
$\begin{eqnarray}
p(\underline{x}) &=& \sum\limits_{i=1}^K \lambda_i \, \mathcal{N}\left(\underline{x}|\underline{\mu}_i,\underline{\underline{\Sigma}}_i\right) \\
\mathcal{N}\left(\underline{x}|\underline{\mu}_i,\underline{\underline{\Sigma}}_i\right) &=&\frac{1}{\sqrt{(2\,\pi)^p\,det \left(\underline{\underline{\Sigma}}_i\right)}}\,\exp\left(-\frac{1}{2}(\underline{x}-\underline{\mu}_i)^T \,\underline{\underline{\Sigma}}_i^{-1} \, (\underline{x}-\underline{\mu}_i) \right)\\
\sum\limits_{i=1}^K \lambda_i &=& 1
\end{eqnarray}$

Hierbei beschreibt $p(\underline{x})$ die Wahrscheinlichkeit (genauer die Wahrscheinlichkeitsdichte) dafür, dass der Zufallsvektor $\underline{x}$ realisiert wird. $p$ ist hierbei die Dimension des Zufallsvektors $\underline{x}$ und $K$ gibt die Anzahl an Komponenten des GMM's an. Wüssten wir von jeder Realisierung, von welcher der Komponenten (einzelnen Normalverteilungen) die jeweilige Realisierung stammt, so könnten wir den Erwartungsvektor $\underline{\mu}$ sowie die Kovarianzmatrix $\underline{\underline{\Sigma}}$ einfach schätzen. Da wir jedoch nicht wissen, von welcher der $K$ Normalverteilungen der Wert realisiert wurde, stellt sich die Schätzung des GMMs als deutlich schwieriger heraus.<br><br>
    Wir werden jetzt ein Verfahren kennenlernen, den so genannten <b>E</b>xpectation <b>M</b>aximization Algorithmus, welches aus den gegebenen Realisierungen der Zufallsvektoren, in anderen Worten unserer Merkmalsvektoren, das zugehörige GMM schätzt.

```python
import numpy as np
import matplotlib as mpl
from sklearn.mixture import GaussianMixture

colors = ["navy", "turquoise"]

def make_ellipses(gmm, ax):
    for n, color in enumerate(colors):
        if gmm.covariance_type == "full":
            covariances = gmm.covariances_[n][:2, :2]
        elif gmm.covariance_type == "tied":
            covariances = gmm.covariances_[:2, :2]
        elif gmm.covariance_type == "diag":
            covariances = np.diag(gmm.covariances_[n][:2])
        elif gmm.covariance_type == "spherical":
            covariances = np.eye(gmm.means_.shape[1]) * gmm.covariances_[n]
        v, w = np.linalg.eigh(covariances)
        u = w[0] / np.linalg.norm(w[0])
        angle = np.arctan2(u[1], u[0])
        angle = 180 * angle / np.pi  # convert to degrees
        v = 2.0 * np.sqrt(2.0) * np.sqrt(v)
        ell = mpl.patches.Ellipse(
            gmm.means_[n, :2], v[0], v[1], 180 + angle, color=color
        )
        ell.set_clip_box(ax.bbox)
        ell.set_alpha(0.5)
        ax.add_artist(ell)
        ax.set_aspect("equal", "datalim")

# load data from file        
X = np.fromfile('data/data-002/oldfaithful.dat',sep=" ").reshape((272,3))
X = X[:,1:3]

# fit gaussian mixture model
gmm = GaussianMixture(n_components=2).fit(X)

# draw result
fig, ax = plt.subplots()
plt.scatter(X[:,0],X[:,1])

for n, color in enumerate(colors):
    covariances = gmm.covariances_[n][:2,:2]
    v, w = np.linalg.eigh(covariances)
    u = w[0] / np.linalg.norm(w[0])
    angle = np.arctan2(u[1], u[0])
    angle = 180 * angle / np.pi  # convert to degrees
    v = 2.0 * np.sqrt(2.0) * np.sqrt(v)
    ell = mpl.patches.Ellipse(
        gmm.means_[n, :2], v[0], v[1], 180 + angle, color=color
    )
    ell.set_clip_box(ax.bbox)
    ell.set_alpha(0.5)
    ax.add_artist(ell)
    #ax.set_aspect("equal", "datalim")
    

plt.show()
```

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.mixture import GaussianMixture
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix 
    
# load dataset
X,y = load_iris(return_X_y=True)

# data splitting (training and test)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.05, random_state=42)


# separate features from class labels
mu_0 = np.mean(X[np.argwhere(y_train==0)],axis=0)
mu_1 = np.mean(X[np.argwhere(y_train==1)],axis=0)
mu_2 = np.mean(X[np.argwhere(y_train==2)],axis=0)
init_val = np.vstack((mu_0,mu_1,mu_2))

gmm = GaussianMixture(n_components=3, means_init=init_val).fit(X_train)

y_pred = gmm.predict(X_test)
print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred)) 
```

Wir wollen uns noch eine weitere Eigenschaft der GMMs anschauen, die eigentlich offensichtlich ist. Ein GMM interpretiert die Merkmalsvektoren als Realisierungen von Zufallsvektoren. Nachdem wir das GMM berechnet haben, können wir doch versuchen, neue Beispiele aus der Verteilung zu generieren. Die Klasse GaussianMixture in sklearn stellt hierzu die Funktion <i>sample</i> zur Verfügung. Wir wollen uns hierzu ein konkretes Beispiel anschauen. Wir verwenden die Gesichtsbilder aus dem Datensatz, den wir mittels der sklearn-Funktion <i>fetch_olivetti_faces</i> extrahieren können (siehe PCA-Beispiel). Um die Dimension zu reduzieren, führen wir zunächst eine PCA durch. Wir verwenden dann die Principal Components um ein GMM zu erstellen. Aus dieser Verteilung ziehen wir neue Beispiele (samples) und rekonstruieren das zugehörige Bild durch die Entwicklung nach den Principal Axes. 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces
from sklearn.decomposition import PCA
from sklearn.mixture import GaussianMixture
from ipywidgets import interactive, fixed
import ipywidgets as widgets

def synthesizeFaces(num_pcs=30, num_components=5):
    # load olivetti dataset
    X,y = fetch_olivetti_faces(return_X_y=True)
    pca = PCA(n_components=num_pcs)
    featureVectors=pca.fit_transform(X)
            
    # create GMM
    gmm = GaussianMixture(num_components).fit(featureVectors)
    XX, y = gmm.sample(n_samples=4)
    
    # transform back to image space
    imgs = pca.inverse_transform(XX).reshape((4,64,64))

    # use colormap gray for plotting images
    plt.set_cmap('gray')

    # print first 4 eigenfaces
    fig, (ax1,ax2,ax3,ax4) = plt.subplots(1,4)
    fig.suptitle("Four samples drawn from the GMM distribution")
    fig.set_size_inches(12,3)

    ax1.set_xticks([])
    ax1.set_yticks([])
    ax2.set_xticks([])
    ax2.set_yticks([])
    ax3.set_xticks([])
    ax3.set_yticks([])
    ax4.set_xticks([])
    ax4.set_yticks([])

    ax1.imshow(imgs[0,:,:])
    ax2.imshow(imgs[1,:,:])
    ax3.imshow(imgs[2,:,:])
    ax4.imshow(imgs[3,:,:])

    plt.show()
    
    # For comparison show all 40 subjects in the database
    subjects = []
    i=0
    while i<400:
        subjects.append(X[i,:].reshape((64,64)))
        i+=10
     
    l=0
    fig, ax = plt.subplots(5,8)
    fig.set_size_inches(30,10)
    fig.suptitle("All 40 subjects in the database")
    for i,ax_row in enumerate(ax):
        for j,axes in enumerate(ax_row):
            axes.set_yticklabels([])
            axes.set_xticklabels([])
            axes.imshow(subjects[l])
            l+=1
    plt.show()        


v=interactive(synthesizeFaces,num_pcs=(20,400),num_components=(1,15))
v  
```
